```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("HandsOnPertemuan3").getOrCreate()

data = [
    ('James', 'Sales', 3000),
    ('Michael', 'Sales', 4600),
    ('Robert', 'Sales', 4100),
    ('Maria', 'Finance', 3000)]
columns = ['EmployeeName', 'Department', 'Salary']

df = spark.createDataFrame(data, schema=columns)
df.show()
```

```python
# TUGAS 1

# Menampilkan kolom tertentu saja
print("=== Nama dan Gaji Karyawan ===")
df.select("EmployeeName", "Salary").show()

# Filter data dengan kondisi (gaji lebih dari 3000)
print("=== Karyawan dengan gaji > 3000 ===")
df.filter(df.Salary > 3000).show()

# Mengelompokkan data berdasarkan departemen dan menghitung rata-rata gaji
print("=== Rata-rata gaji per departemen ===")
df.groupBy("Department").avg("Salary").show()

# Menghitung gaji maksimum per departemen
print("=== Gaji maksimum per departemen ===")
df.groupBy("Department").max("Salary").show()

# Mengurutkan data berdasarkan gaji tertinggi
print("=== Urutan gaji tertinggi ===")
df.orderBy(df.Salary.desc()).show()
```

```python
# Contoh operasi transformasi DataFrame
df.select('EmployeeName', 'Salary').show()
df.filter(df['Salary'] > 3000).show()
df.groupBy('Department').avg('Salary').show()
```

```python
# TUGAS 2

# 1. Menampilkan hanya kolom tertentu (select)
print("=== Menampilkan Nama dan Gaji ===")
df.select("EmployeeName", "Salary").show()

# 2. Filter data dengan kondisi (gaji > 3000)
print("=== Karyawan dengan gaji > 3000 ===")
df.filter(df["Salary"] > 3000).show()

# 3. Hitung rata-rata gaji per departemen
print("=== Rata-rata gaji per departemen ===")
df.groupBy("Department").avg("Salary").show()

# 4. Hitung gaji maksimum per departemen
print("=== Gaji maksimum per departemen ===")
df.groupBy("Department").max("Salary").show()

# 5. Hitung total gaji (sum) per departemen
print("=== Total gaji per departemen ===")
df.groupBy("Department").sum("Salary").show()
```

```python
# TUGAS 3

# 1. Menambahkan kolom baru SalaryBonus = 10% dari Salary
print("=== Menambahkan kolom SalaryBonus (10% dari gaji) ===")
df = df.withColumn("SalaryBonus", df["Salary"] * 0.1)
df.show()

# 2. Menambahkan kolom baru TotalCompensation = Salary + SalaryBonus
print("=== Menambahkan kolom TotalCompensation ===")
df = df.withColumn("TotalCompensation", df["Salary"] + df["SalaryBonus"])
df.show()
```

```python
# Contoh menggunakan window functions
from pyspark.sql.window import Window
from pyspark.sql import functions as F

windowSpec = Window.partitionBy('Department').orderBy('Salary')
df.withColumn('Rank', F.rank().over(windowSpec)).show()
```

```python
# TUGAS 4

# Contoh dataset
data = [
    ("A", "IT", 3000),
    ("B", "IT", 4000),
    ("C", "HR", 2500),
    ("D", "HR", 2800),
    ("E", "Finance", 3500),
    ("F", "Finance", 4500),
    ("G", "IT", 5000)
]

columns = ["Name", "Department", "Salary"]

# Membuat DataFrame
df = spark.createDataFrame(data, columns)

print("=== Data Awal ===")
df.show()

# Definisi window untuk perhitungan rank dalam tiap department
windowSpec = Window.partitionBy("Department").orderBy(F.col("Salary").desc())

# Menambahkan kolom ranking
df_rank = df.withColumn("Rank", F.rank().over(windowSpec))

print("=== Ranking berdasarkan Salary tiap Department ===")
df_rank.show()

# Definisi window untuk running total (cumulative sum salary per department)
windowSpecRunning = Window.partitionBy("Department").orderBy("Salary").rowsBetween(Window.unboundedPreceding, Window.currentRow)

# Menambahkan kolom Running Total
df_running = df.withColumn("RunningTotal", F.sum("Salary").over(windowSpecRunning))

print("=== Running Total Salary tiap Department ===")
df_running.show()


```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# 1. Buat SparkSession
spark = SparkSession.builder.appName("Tugas5").getOrCreate()

# 2. Load dataset CSV
df = spark.read.csv("/home/bila/Downloads/Reviews.csv", header=True, inferSchema=True)

# 3. Mengecek jumlah baris dan kolom
print("Jumlah baris:", df.count())
print("Jumlah kolom:", len(df.columns))

# 4. Mengecek schema (struktur dataset)
print("=== Schema Dataset ===")
df.printSchema()

# 5. Eksplorasi dasar: jumlah review per skor
print("=== Jumlah Review per Rating (Score) ===")
df.groupBy("Score") \
    .count() \
    .orderBy("Score") \
    .withColumnRenamed("count", "review_count") \
    .show(truncate=False)

# 6. Top 10 produk dengan review terbanyak
print("=== Top 10 Produk dengan Review Terbanyak ===")
df.groupBy("ProductId") \
    .count() \
    .orderBy(F.desc("count")) \
    .withColumnRenamed("count", "review_count") \
    .show(10, truncate=False)

# 7. Statistik deskriptif pada kolom Score
print("=== Statistik Deskriptif Rating (Score) ===")
df.select("Score").describe().show(truncate=False)

# 8. Filter review dengan rating tinggi (>4)
print("=== Review dengan Rating > 4 ===")
df.filter(F.col("Score") > 4) \
    .select("ProductId", "UserId", "Score", "Summary") \
    .show(10, truncate=False)
```
